```python
seoul_school_df = pd.read_excel('./data/elem_school_folium.xlsx')
seoul_school_df.head()
```

```python
# Feature Selection
# , , , , 

school_subset_df = seoul_school_df.filter(['', '', '', '', ''])
school_subset_df.head()
```

```python
#
school_subset_df.isnull().sum()
school_subset_df.dropna(inplace=True)
school_subset_df.isnull().sum()
```

```python
#  :     
school_subset_df[(school_subset_df[''] == '') &
                 (school_subset_df[''].str.contains(''))]

map_df = school_subset_df[ school_subset_df[''].str.contains('') &
                 school_subset_df[''].str.contains('') ]
```
