```python
# 각 종별로 가장 큰 값과 가장 작은 값의 비율을 구한다면?
# agg()

def get_ratio(x):
    return x.max() / x.min()
```

```python
# 종별로 가장 긴 꽃잎길이가 큰 3개의 데이터를 뽑아낸다면?
# sort_values() : 
iris_datasets.sort_values(by='petal_length' , ascending=False)[:3]
```

```python
def max3_petal_length(df) :
    return df.sort_values(by='petal_length' , ascending=False)[:3]
```

```python
# iris_datasets.groupby(iris_datasets.species).agg(max3_petal_length) # agg쓰면 valueerror 발생..따라서 apply를 써야한다.

iris_datasets.groupby(iris_datasets.species).apply(max3_petal_length)
```

- transform() : 그룹별 계산은 동일하지만 데이터프레임 자체를 변화
- 꽃잎길이가 해당 종 내에서 대/중/소 어느것에 해당되는지에 대한 프레임을 만들고 싶다면?
- cut() : 동일 길이로 나누어서 범주를 만들고 그룹에 대한 통계량
- qcut() : 동일 갯수로 나누어서 범주를 만들고 그룹에 대한 통계량

```python
iris_datasets.groupby(iris_datasets.species)
```

```python
def cat3_petal_length(series) :
    return pd.qcut(series, 3, labels=['small', 'medium', 'large']).astype(str)
```

```python
iris_datasets['category'] = iris_datasets.groupby(iris_datasets.species).petal_length.transform(cat3_petal_length)
iris_datasets['category'] = iris_datasets.groupby(iris_datasets.species).petal_length.apply(cat3_petal_length)
# iris_datasets['category'] = iris_datasets.groupby(iris_datasets.species).petal_length.agg(cat3_petal_length) # error 발생
iris_datasets.head()
```

- filter() : 그룹 객체에 filter() 적용할 때 조건식을 자진 함수를 전달하면 조건이 참인 그룹만 남기게 된다.

```python
titanic_datasets = sns.load_dataset('titanic')
```

```python
titanic_datasets.head()
```

```python
grp_class_titanic = titanic_datasets.groupby(['class'])
grp_class_titanic
```

```python
# 데이터의 갯수가 200개 이상인 그룹만 필터링한다면?
#1
def countOver(df):
    return len(df) >= 200

grp_class_titanic.filter(countOver)

#2
# 람다 식으로 만들어보기
grp_filter = grp_class_titanic.filter(lambda x : len(x) >= 200)
grp_filter
```

```python
# Age 열의 평균이 30보다 작은 그룹만 필터링
#1
def avgUnder(df):
    return df.age.mean() < 30

grp_class_titanic.filter(avgUnder)

#2
grp_filter = grp_class_titanic.filter(lambda x : x.age.mean() < 30)
grp_filter
```

```python
# 반복문을 이용하여 출력을 한다면?

grp_filter.loc[0]

for idx in grp_filter.index :
    print(grp_filter.loc[idx])
    print('\n')

        
```

```python
tips_datasets = sns.load_dataset('tips')
tips_datasets
```

```python
# 식사대금 대비 팁의 비율이 어떤 경우에 가장 높은지를 찾는다면?

# feature 추가
tips_datasets['ratio'] = tips_datasets['tip'] / tips_datasets['total_bill']

tips_datasets
```

```python
# 성별로 grouping
tips_datasets.groupby('sex').size()
```

```python
# 성별과 흡연유무로 grouping
tips_datasets.groupby(['sex', 'smoker']).size()
```

```python
# 성별과 흡연유무로 grouping
tips_datasets.groupby(['sex', 'smoker']).count()
```

```python
# 성별과 흡연유무로 grouping
pd.pivot_table(data   = tips_datasets, 
               index  = 'sex',
               columns= 'smoker',
               values = 'ratio', # 이 열을 사용하겠다.
               aggfunc= 'count',
               margins= True # 카운트되는 행과 열에 대한 집계가 출력
```

```python
# 성별과 흡연여부에 따른 평균 팁 비율을 살펴본다면?

pd.pivot_table(data = tips_datasets, 
               index = 'sex',
               columns='smoker',
               values ='ratio',
               aggfunc='mean')
```

```python
# groupby 활용
# series 타입
tips_datasets.groupby(['sex', 'smoker'])['ratio'].mean()
# df 타입
tips_datasets.groupby(['sex', 'smoker'])[['ratio']].mean()
```

```python
# 나이에 대한 범주
titanic_datasets['age_qcut'] = pd.qcut(titanic_datasets['age'], 3, labels=['A', 'B', 'C']).astype(str)
titanic_datasets
```

```python
# 성별, 선실, 나이 그룹(age_qcut)에 의한 생존율을 데이터프레임으로 생성하고
# 행에는 성별 및 나이그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.


resultDF = titanic_datasets.pivot_table(index = ['sex', 'age_qcut'], 
                             columns = 'pclass',
                             values = 'survived',
                             margins = True,
                             margins_name = 'survived_ratio',
                             aggfunc = 'count')
```

```python
resultDF
```

```python
all_cnt = resultDF.loc['survived_ratio', 'survived_ratio'][0]
```

```python
survived_ratio_df = resultDF/ all_cnt
```
