```python
# pip install pymongo
import pymongo as mongo
```

- 1. connection mongodb

```python
# local
conn = mongo.MongoClient()
# remote
# conn = mongo.MongoClient(server , 27017)
```

- 2. Database create

```python
articleDB = conn.articleDB
print(articleDB)
```

```python
articleCollection = articleDB['article']
print(articleCollection)
```

- Document Insert
- -insert_one()  - insertOne()(mongo shell)
- -insert_many() - insertMany()(mongo shell)

```python
doc01 = {'author' : 'jslim' , 
         'subject' : '드디어 몽고' , 
         'content' : '잉~~ 쉽네' , 
         'tags' : ['mongobd' , 'python' , 'pymongo'] , 
         'view' : 50 } 
```

```python
articleCollection.insert_one(doc01)
```

- Document find
- find_one()
- find()

```python
articleCollection.find_one()
```

```python
doc02 = {'author' : 'campus' , 
         'subject' : '인공지능' , 
         'content' : '비정형 데이터 학습' , 
         'tags' : ['mongobd' , 'python' , 'pymongo'] , 
         'view' : 100 } 
```

```python
articleCollection.insert_one(doc02)
```

```python
articleCollection.find_one( {'author' : 'campus' } )
```

```python
cursor = articleCollection.find()
for doc in cursor : 
    print(doc)
```

```python
doc03 = {'author' : 'jslim' , 
         'subject' : '드디어 몽고' , 
         'content' : '잉~~ 쉽네' , 
         'tags' : ['mongobd' , 'pymongo'] , 
         'view' : 200 } 
```

```python
articleCollection.insert_one(doc03)
```

```python
cursor = articleCollection.find( {'author' : 'jslim' ,  'view' : {'$gte' : 200} } )
for doc in cursor : 
    print(doc)
```

```python
cursor = articleCollection.find( {'author' : 'jslim' ,  'view' : {'$gte' : 200} } , {'author' : 1 , '_id' : 0 } )
for doc in cursor : 
    print(doc)
```

```python
# update
articleCollection.update_one({'author' : 'campus'} , {'$set' : {'view' : 150}} )
```

```python
cursor = articleCollection.find()
for doc in cursor : 
    print(doc)
```

```python
# remove
# delete_one()
# delete_many()
articleCollection.delete_one( {'author' : 'campus'})
```

```python
cursor = articleCollection.find()
for doc in cursor : 
    print(doc)
```

```python
articleCollection.delete_many( {'author' : 'jslim'})
```

```python
cursor = articleCollection.find()
for doc in cursor : 
    print(doc)
```
