```python
map = f.Map(location=[53.3493, -6.2444], zoom_start=12, tiles='Stamen Toner')

marker = f.Marker([53.3493, -6.2444],
                 popup='we',
                 icon = f.Icon(color='blue'))
marker.add_to(map)
map
```

```python
marker = f.CircleMarker([53.3493, -6.2444],
                 popup='we',
                 fill_color='pink',
                color='skyblue')
marker.add_to(map)
map
```

```python
# 서울지역 대학교 위치 정보를 읽어보자

seoul_univ_df = pd.read_excel('./data/서울지역 대학교 위치.xlsx')
# seoul_univ_df.shape[0]
# len(seoul_univ_df)

# Marker 표시
map = f.Map(location=[37.566652, 126.978449], zoom_start=11.5, tiles='Stamen Toner')

for idx in range(seoul_univ_df.shape[0]):

    marker = f.Marker(location=[ seoul_univ_df.iloc[idx,[1]], seoul_univ_df.iloc[idx,[2]] ], 
                     popup=seoul_univ_df.iloc[idx,[0]], icon = f.Icon(color='pink'))
    marker.add_to(map)
map

# for name, lat, lng in zip(seoul_univ_df.index, seoul_univ_df.위도, seoul_univ_df.경도):
#     f.Marker([lat, lng], popup=name).add_to(map)
# map
```
