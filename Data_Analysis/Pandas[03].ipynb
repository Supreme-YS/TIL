### apply 함수
- lambda 식과 같이 사용한다.
- 행이나 열 단위로 복잡한 데이터 가공이 필요한 경우
- apply 함수는 인자로 함수를 넘겨 받을 수 있다

```python
def get_square(a) :
    return a ** 2
```

```python
print('제곱근 : ', get_square(2) )
```

```python
# 위 코드를 람다식으로 바꾼다면?
# lambda 인자 : 조건부
# 람다식을 사용하여 함수처럼 사용이 가능하다!
lambda_square = lambda a : a ** 2
```

```python
print('제곱근 : ', lambda_square(2) )
```

```python
sampleDF = pd.DataFrame(np.random.randint(0, 10, (6,4) ))
```

```python
sampleDF.columns = ['A', 'B', 'C', 'D'] # 칼럼명 바꾸기
sampleDF.index   = pd.date_range('20210302', periods=6) # 로우명 바꾸기
```

```python
sampleDF
```

```python
# 각 행의 column에 대해서 최대값 - 최소값을 구해서 새로운 columns 추가?
func = lambda x : x.max() - x.min()
```

```python
sampleDF['row max-min'] = sampleDF.apply(func, axis = 1)
```

```python
# embark_town 의 글자수를 새로운 칼럼 embark_len 으로 만들고 싶다면?
# Series가 들어가면 축은 의미가 없다
# DF이 들어갔을 때 축 명시 필요하다

datasets['embark_len'] = datasets['embark_town'].apply(lambda x : len(str(x)) )
```

- lambda if ~ else 구문식
- lambda 매개변수 : trun if 조건식 else false

```python
age_func = lambda x : 'child' if x < 20 else 'adult'
```

```python
# age 피처를 이용하여 나이가 20세 미만이면 Child, 그렇지 않으면 Adult 로 구분하여 child_adult 컬럼을 추가한다면?

datasets['child_adult'] = datasets['age'].apply(age_func)
```

```python
datasets[['age', 'child_adult']]
```

```python
# data frame이 매개변수로 들어오는 상황! 어떻게 처리해야 할까?

datasets['child_adult_02'] = datasets.apply(lambda df : 'child' if df['age'] < 20 else 'adult', axis = 1)
# datasets['child_adult_02'] = datasets.apply(lambda df : 'child' if df.age < 20 else 'adult', axis = 1)

datasets
```

```python
# 15세 이하는 child ,  15~60 까지는 adult , 61 이상은 elderly 로 분류를해서 age_division 컬럼에 저장
# lambda 식으로 apply함수를 이용해서 구현한다면?


new_age_func = lambda x : 'child' if x < 15 else 'adult' if 15 < x < 60 else 'elderly'
```

```python
datasets['age_division'] = datasets['age'].apply(new_age_func)
```
