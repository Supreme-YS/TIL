### apply 함수
- lambda 식과 같이 사용한다.

```python
def get_square(a) :
    return a ** 2
```

```python
print('제곱근 : ', get_square(2) )
```

```python
# 위 코드를 람다식으로 바꾼다면?
# lambda 인자 : 조건부
# 람다식을 사용하여 함수처럼 사용이 가능하다!
lambda_square = lambda a : a ** 2
```

```python
def get_age_category(age) :
    
    category = ""
    
    if age <= 5 :
        category = "baby"
    elif age <=12 :
        category = "child"
    elif age <=19 :
        category = "teenager"
    elif age <= 24 :
        category = "student"
    elif age <= 39 :
        category = "young adult"
    elif age <= 60 :
        category = "adult"
    else :
        category = "elderly"
    
    return category
```

```python
# 함수를 호출하기
datasets['age_category'] = datasets['age'].apply(lambda age : get_age_category(age))
```

```python
datasets['age_category'].value_counts()
```

### Frame Merge
- 데이터 프레임 합치기
- merge()
- concat()
- join()

```python
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
```

```python
df1
```

```python
df2
```

```python
# concat 열의 이름을 합집합 형식으로 해서 하나의 DF로 만들어진다.
result1 = pd.concat([df1, df2])
result1
```

```python
# ignore_index = True
# 합쳐지지만 index가 순차적으로 입력이 된다.
result2 = pd.concat([df1, df2], ignore_index = True)
result2
```

```python
result3 = pd.concat([df1, df2], axis = 1)
result3
```

```python
result4 = pd.concat([df1, df2], axis = 1, join='inner') # 교집합의 옵션을 적용하겠다.
result4
```

```python
result5 = pd.concat([df1, df2], axis = 1, join='inner', ignore_index=True) # 교집합의 옵션을 적용하겠다.
result5
```

```python
data1 = {
    '학번' : [1,2,3,4],
    '이름' : ['섭섭해' , '김한준' , '김선림' , '최호진'],
    '학년' : [2,4,1,3]
}

data2 = {
    '학번' : [1,2,4,5],
    '학과' : ['CS' , 'MATH' , 'MATH' , 'CS'],
    '학점' : [2.4, 4.5 , 3.4, 4.2]
}
```

```python
stuDF = pd.DataFrame(data1)
```

```python
majorDF = pd.DataFrame(data2)
```

```python
stuDF
```

```python
majorDF
```

```python
# inner join과 유사하다.
# on = key
# how = inner | left | right | outer
pd.merge(stuDF, majorDF)
```

```python
pd.merge(stuDF, majorDF, on='학번', how='inner') #on 은 기준점, how 는 조인 형태 설정
```
