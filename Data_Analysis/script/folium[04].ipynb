```python
# 서울지역 대학교 위치 정보를 읽어보자

seoul_uni_df = pd.read_excel('./data/서울지역 대학교 위치.xlsx')
# seoul_uni_df.shape[0]
# len(seoul_uni_df)

map = f.Map(location=[37.566652, 126.978449] , zoom_start=12)

# Marker 표시
for name, lat, lng in zip(seoul_uni_df.index, seoul_uni_df.위도, seoul_uni_df.경도):
    f.Marker([lat, lng], popup=name).add_to(map)

map
```

```python
seoul_school_df = pd.read_excel('./data/elem_school_folium.xlsx')
seoul_school_df.head()
```

```python
# Feature Selection
# 학교명, 학교급구분, 소재지도로명주소, 위도, 경도
school_subset_df = seoul_school_df.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
school_subset_df.head()
```

```python
# 결측값 확인, 처리
school_subset_df.isnull().sum()
```

```python
school_subset_df.dropna(inplace = True)
school_subset_df.isnull().sum()
```

```python
# 소재지도로명주소 서울을 포함하고 학교급구분이 초등학교만 추출
map_df = school_subset_df[ school_subset_df['소재지도로명주소'].str.contains('서울특별시') &
                  school_subset_df['학교급구분'].str.contains('초등학교')]
```

```python
map_df
```

```python
# 지도 시각화
seoul_ele_map = f.Map(location=[37.55 , 126.98] , zoom_start=12 , tiles='Stamen Terrain')
for idx in map_df.index :
    f.CircleMarker([ map_df.loc[idx , '위도'], map_df.loc[idx , '경도']] , 
                     popup = map_df.loc[idx , '학교명'] , 
                     color = 'brown' , # 원의 둘레 색상
                     fill_color = 'coral' , # 원을 채우는 색
                     fill = True , 
                     fill_opacity= .2 ).add_to(seoul_ele_map)
    
#     f.Marker([ map_df.loc[idx , '위도'], map_df.loc[idx , '경도']] , 
#              popup = map_df.loc[idx , '학교명'] , 
#              icon = f.Icon(color='red')).add_to(seoul_ele_map)
    
        

seoul_ele_map
```
